```python
device = torch.device("cuda:0" if torch.cuda.is_available() else 'cpu')
```

```python
class MyDataset(Dataset):  
    def __init__(self, data, target, transform=None, flatten=False):
        if flatten:
            data = np.reshape(data, [-1, 28*28])
        self.data = torch.from_numpy(data).float()
        self.target = torch.from_numpy(target).long()
        self.transform = transform
    
    def __getitem__(self, index):
        x = self.data[index]
        y = self.target[index]
    
        if self.transform:
            x = self.transform(x)      
        return x,y
  
    def __len__(self):
        return len(self.data)
```

```python
class Network(nn.Module):
    def __init__(self, n_hidden):
        super().__init__()
        
        self.hidden = nn.Linear(784, n_hidden)
        self.output = nn.Linear(n_hidden, 10)
        self.sigmoid = nn.Sigmoid()
        self.softmax = nn.Softmax(dim=1)
        
    def forward(self, x):
        x = self.hidden(x)
        x = self.sigmoid(x)
        x = self.output(x)
        x = self.softmax(x) 
        return x
```

```python
transform=transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])
```

```python
model = Network(32).to(device)
```

```python
def test(model, device, test_loader):
    model.eval()
    test_loss = 0
    correct = 0
    with torch.no_grad():
        for data, target in test_loader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += criterion(output, target).item()
            pred = output.argmax(dim=1, keepdim=True)
            correct += pred.eq(target.view_as(pred)).sum().item()

    test_loss /= len(test_loader.dataset)

    return 100. * correct / len(test_loader.dataset), test_loss
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
x_train = np.reshape(x_train, (len(x_train), 1, 28, 28))/255
x_test = np.reshape(x_test, (len(x_test), 1, 28, 28))/255
```

```python
test_dataset = MyDataset(x_test, y_test, transform=None, flatten=True)
test_loader = DataLoader(test_dataset, batch_size = 512, shuffle=False)
```

```python
criterion = nn.CrossEntropyLoss()
```

```python
def darwin(weight1, bias1, weight2, bias2):
    model = Network(32).to(device)
    
    with torch.no_grad():
        model.hidden.weight = nn.Parameter(torch.from_numpy(weight1).float())
        model.hidden.bias = nn.Parameter(torch.from_numpy(bias1).float())
        
        model.output.weight = nn.Parameter(torch.from_numpy(weight2).float())
        model.output.bias = nn.Parameter(torch.from_numpy(bias2).float())
    
    return test(model, device, test_loader)
```

```python
solutions = []
for s in range(1000):
    solutions.append((np.random.randn(32, 784),
                      np.random.randn(32),
                      np.random.randn(10, 32),
                      np.random.randn(10)
                     ))
```

```python
def sort_tuple(tup): 
    tup.sort(key = lambda x: x[0]) 
    return tup 
```

```python
best_solutions = []
worst_solutions = []

for i in range(100):
    ranked_solutions = []
    for s in solutions:
        ranked_solutions.append((darwin(s[0], s[1], s[2], s[3]), s))
    ranked_solutions = sort_tuple(ranked_solutions)

    print(f"Generation: #{i} the best offspring")
    print(ranked_solutions[-1][0])
    print(f"Generation: #{i} the worst offspring")
    print(ranked_solutions[0][0])
    
    best_solutions.append(ranked_solutions[-1][0])
    worst_solutions.append(ranked_solutions[0][0])

    elements = []
    for s in ranked_solutions[-100:]:
        elements.append(s[1])

    new_generation = []

    for _ in range(1000):
        e1 = random.choice(elements)[0] * random.uniform(0.9, 1.1)
        e2 = random.choice(elements)[1] * random.uniform(0.9, 1.1)
        e3 = random.choice(elements)[2] * random.uniform(0.9, 1.1)
        e4 = random.choice(elements)[3] * random.uniform(0.9, 1.1)

        new_generation.append((e1, e2, e3, e4))

    solutions = new_generation
```

```python
plt.plot(best_solutions)
plt.plot(worst_solutions)
plt.xlabel('Generation')
plt.ylabel('Accuracy')
```
